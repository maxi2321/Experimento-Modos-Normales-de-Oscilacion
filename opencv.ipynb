```python
import cv2
import matplotlib.pyplot as plt

# Ruta al video
video_path 

# Abrir el video
cap = cv2.VideoCapture(video_path)

# Obtener FPS y cantidad total de frames
fps = cap.get(cv2.CAP_PROP_FPS)
total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

# Generar lista de tiempos (en segundos)
tiempos = [i / fps for i in range(total_frames)]

# Graficar
plt.figure(figsize=(10, 4))
plt.plot(tiempos, marker='o', linestyle='-', color='blue')
plt.xlabel('Número de frame')
plt.ylabel('Tiempo [s]')
plt.title(f'Tiempo vs Frame (FPS = {fps:.2f})')
plt.grid(True)
plt.tight_layout()
plt.show()

cap.release()
```

```python
import cv2
import os

# Carpeta donde guardar los frames
output_folder = 'frames_guardados'
os.makedirs(output_folder, exist_ok=True)

# Abrir el video
cap = cv2.VideoCapture(video_path)

frame_num = 0

while True:
    ret, frame = cap.read()
    if not ret:
        break  # Se terminó el video

    # Nombre del archivo (ej. frame_0001.png)
    filename = os.path.join(output_folder, f'frame_{frame_num:04d}.png')

    # Guardar el frame como imagen PNG
    cv2.imwrite(filename, frame)

    frame_num += 1

cap.release()
print(f"Se guardaron {frame_num} frames en la carpeta '{output_folder}'.")
```
