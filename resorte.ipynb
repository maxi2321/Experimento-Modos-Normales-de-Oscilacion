# Resortes metalicos

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Datos en gramos y milímetros
masas_g = np.array([0, 100, 50, 500, 200, 400, 250, 700, 300, 220])
elongaciones_mm = np.array([56, 73, 65, 136, 89, 121, 97, 167, 105, 93])

# Conversión a SI
masas = masas_g / 1000  # kg
elongaciones = elongaciones_mm / 1000  # m

# Gravedad local en Salta
g = 9.78
fuerzas = masas * g  # N

# Ajuste lineal F = kx
X = elongaciones.reshape(-1, 1)
y = fuerzas
modelo = LinearRegression()
modelo.fit(X, y)

k = modelo.coef_[0]
y_pred = modelo.predict(X)

# Calcular el error estándar de k
residuos = y - y_pred
n = len(X)
s_squared = np.sum(residuos**2) / (n - 2)
x_mean = np.mean(X)
Sxx = np.sum((X - x_mean)**2)
error_k = np.sqrt(s_squared / Sxx)

# Redondear el error a 1 cifra significativa
potencia = int(np.floor(np.log10(abs(error_k))))
error_redondeado = round(error_k, -potencia)
# Redondear k al mismo orden
k_redondeado = round(k, -potencia)

# Mostrar resultados
print(f"Constante elástica: k = {k_redondeado} ± {error_redondeado} N/m")

# Gráfico
plt.scatter(X, y, label='Datos experimentales')
plt.plot(X, y_pred, color='red', label=f'k = {k_redondeado} ± {error_redondeado} N/m')
plt.xlabel('Elongación (m)')
plt.ylabel('Fuerza (N)')
plt.title('Resorte morado')
plt.grid(True)
plt.legend()
plt.show()
```
