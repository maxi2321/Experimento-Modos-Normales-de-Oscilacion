```python
import pandas as pd
import matplotlib.pyplot as plt

# === Configuracin ===
archivo_csv = '/Users/admin/Documents/Fisica experimental/Experimento-Modos-Normales-de-Oscilacion/MasasIgualesPlasticosaModoLento/trayectorias.csv'
t_max = 10       # Tiempo mximo a graficar en segundos
Fsize = 28        # Tamao de fuente

# === Cargar datos ===
df = pd.read_csv(archivo_csv)

# === Filtrar hasta t_max ===
df = df[df['tiempo_s'] <= t_max]

# === Extraer y centrar trayectorias ===
t = df['tiempo_s']
x1 = df['x_rojo'] - df['x_rojo'].mean()
x2 = df['x_azul'] - df['x_azul'].mean()

# === Grfico ===
plt.figure(figsize=(20, 6))
plt.plot(t, x1, 'r.-', label='$m_1$')
plt.plot(t, x2, 'b.-', label='$m_2$')
plt.xlabel('t [s]', fontsize=Fsize)
plt.ylabel('x [Px]', fontsize=Fsize)
plt.xticks(fontsize=Fsize)
plt.yticks(fontsize=Fsize)
plt.tight_layout()
plt.show()
```
